# 日期格式数据处理

```python
# 对购买日期进行转换
df['buy_date'] = pd.to_datetime(df['day'],format='%Y%m%d',errors = 'coerce')
#加errors防止报错
```

```python
# 可以提取对应年月日,必须为timestamp才可以，具体的时间点
df['buy_date'].dt.year.head(5)
```

时间差

```python
# 对时间进行计算
df['diff_day'] =  pd.datetime.now() - df['buy_date']
df.info()
df.head(5)
```

```python
# 时间差由天数，小时和毫秒构成
# timedelta可以理解为时间差类型
df['diff_day'].dt.days# 提取天数
df['diff_day'].dt.seconds# 提取秒数
df['diff_day'].dt.microseconds# 提取毫秒数
```

```python
pd.Timedelta?
```

```python
# 可以使用pd.Timedelta进行转换
#转换为天数
df['时间差'] = df['diff_day']/pd.Timedelta('1 D')
df.head(5)
```

```python
pd.Timedelta('1 D')
```

```python
#转换为分钟
df['时间差'] = df['diff_day']/pd.Timedelta('1 M')
df.head(5)
```

```python
# 显示为3小数
df['时间差'].round(decimals=3).head(5)
```

```python
# 这种方式也可以，M  代表月份数,D代表天数,Y代表年份
df['diff_day'].astype('timedelta64[Y]').head(5)
```

# 高阶函数数据处理

```python
#婴儿信息
df2 = pd.read_csv('sam_tianchi_mum_baby.csv',encoding = 'utf-8',dtype =str)
df2.head(5)
```

```python
def f(x):
    if '0' in x:
        return '女'
    elif '1' in x :
        return '男'
    else:
        return '未知'
```

```python
# 0代表女，1代表男，2代表未知
df2['性别'] = df2['gender'].apply(f)
```

```python
df2.head(5)
```

```python
#使用map函数
df2['性别'] = df2['gender'].map({'0':'女','1':'男','2':'未知'})
```

```python
#结合lambda替换
df2['user_id'].apply(lambda x: x.replace(x[1:3],'**')).head(5)
```

```python
#提取年份
df2['birthday'].apply(lambda x: str(x)[0:4]).head(5)
```
